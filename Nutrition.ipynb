```python
# Open and read HTML document with Beautiful Soup
fd = open("D:/Statistics/540/Data/Human_nutrition.html", "r", encoding="utf8")
soup = BeautifulSoup(fd)
fd.close()

# Display contents
print(soup.prettify())
```

```python
# Get all tables in HTML document
tables = soup.findAll("table")

print("Number of tables = " + str(len(tables)))
```

```python
captions = [caption for caption in tables[0].findAll('thead')]

for caption in captions:
    header = caption.findAll('tr')
    for h in header:
        output = h.find('th')
        #print(str(output.contents).replace("'","").replace("\n",""))
        if len(output) > 0:
            print(str(output[0].contents).strip())
```
